```python
#splitting data to train and test splits
X_train, X_test, y_train, y_test = train_test_split(scaled_inputs, y2, test_size=0.2, random_state=42)
X_train
print(y_test)
```

```python
# plotting n_estimators

score_array=[]
for estimator_value in range(250,320,10):
    model = GradientBoostingClassifier(n_estimators=estimator_value,learning_rate=0.1,max_depth=1,random_state=0).fit(X_train, y_train)
    score_array.append(model.score(X_test, y_test))

plt.xlabel('number of estimators')
plt.ylabel('accuracy score')
plt.plot(np.arange(250,320,10),score_array,"o-")
```

```python
n_estimators_gbc = GradientBoostingClassifier(n_estimators=300,learning_rate=0.1,max_depth=1,random_state=0).fit(X_train, y_train)
n_estimators_gbc.score(X_test, y_test)
```

```python
# plotting learning_rate

score_array=[]
for learning_value in np.arange(0.75,1.5,0.05):
    model = GradientBoostingClassifier(n_estimators=100,learning_rate=learning_value,max_depth=1,random_state=0).fit(X_train, y_train)
    score_array.append(model.score(X_test, y_test))

plt.xlabel('learning rate')
plt.ylabel('score')
plt.plot(np.arange(0.75,1.5,0.05),score_array,"o-")
```

```python
learning_rate_gbc = GradientBoostingClassifier(n_estimators=100,learning_rate=1.25,max_depth=1,random_state=0).fit(X_train, y_train)
learning_rate_gbc.score(X_test, y_test)
```
